# Customer Spending Analysis

## Overview
This project analyzes customer spending patterns based on income levels using K-means clustering. It aims to segment customers into distinct groups to better understand their spending behaviors.
## Features
Data loading and preprocessing using Pandas
Exploratory Data Analysis (EDA)
K-means clustering implementation
Elbow method for optimal cluster determination
Visualization of clusters and centroids
Requirements
Python 3.x
pandas
numpy
matplotlib
scikit-learn

## Data
The script uses a CSV file named 'Customer spent analysis.csv' with the following columns:
Income
Spend
## Methodology
Load and preprocess the data
Perform basic EDA
Use the elbow method to determine the optimal number of clusters
Apply K-means clustering
Visualize the results
Output
The script generates two plots:
Elbow curve to determine the optimal number of clusters
Scatter plot showing the clustered data points and centroids
## Results
The analysis segments customers into 4 distinct groups based on their income and spending patterns. This segmentation can be used for targeted marketing strategies or personalized customer experiences.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('D:\\python_lab_20004\\test\\Customer spent analysis.csv')
df
```

```python
print(df.shape)
print(df.describe())
print(df.head(5))
```

```python
Income=df['Income'].values
Spend=df['Spend'].values
X=np.array(list(zip(Income,Spend)))
X
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,random_state=0)
    km.fit(X)
    wcss.append(km.inertia_)
plt.plot(range(1,11),wcss,color='red',marker='8')
plt.title('Optimal K value')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model=KMeans(n_clusters=4,random_state=0)
y_means=model.fit_predict(X)
```

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1],s=50,c='yellow',label='1')
plt.scatter(X[y_means==1,0],X[y_means==1,1],s=50,c='blue',label='2')
plt.scatter(X[y_means==2,0],X[y_means==2,1],s=50,c='green',label='3')
plt.scatter(X[y_means==3,0],X[y_means==3,1],s=50,c='red',label='4')

plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=100,marker='s',c='black')
plt.title('Income Spend Analysis')
plt.xlabel('Income')
plt.ylabel('Spend')
plt.legend()
plt.show()
```
